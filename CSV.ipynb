```python
from faker import Faker
import csv
```

```python
output=open('data.csv','w')
faker=Faker()
```

```python
header=['name','agent','street','city','state','zip','lng','lat']
```

```python
mywriter=csv.writer(output)
mywriter.writerow(header)
```

```python
for r in range(1000):
    mywriter.writerow([faker.name(),faker.random_int(min=18,max=80,step=1),faker.street_address(),faker.city() \
                       ,faker.state(),faker.zipcode(),faker.longitude(),faker.latitude()])
output.close()
```
